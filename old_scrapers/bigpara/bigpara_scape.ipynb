```python
from urllib.request import urlopen, Request
from bs4 import BeautifulSoup
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
from new_class import News
SOURCE = "BIGPARA"
```

```python
bigpara_base_url = "https://bigpara.hurriyet.com.tr"
bigpara_news_url = bigpara_base_url + "/borsa/haber/"
```

```python
# fill the title,source,news_url, date fields of the news, then extract the content from the given url
all_news = []
# 40 pages
for i in range(40):
    url = bigpara_news_url + str(i+1) + "/"
    req = Request(url, headers={'User-Agent': 'Mozilla/5.0'})
    html = urlopen(req).read()
    soup = BeautifulSoup(html, 'html.parser')
    newsTable = soup.find_all(class_="tBody")
    newslist = newsTable[0].find_all("ul")
    # print("Page: ", i+1)
    # print("News Count: ", len(news))
    # print("News: ", news[0])
    print("Page: ", i+1)
    print("news count: ", len(newslist))
    for new in newslist:
        # find the title in news html which has tag with h2
        new_title = new.find("h2").text
        new_url = bigpara_base_url + new.find("a")["href"]
        # date is kept in this html element : <li class="cell012 tar fsn">23.10.2024</li>
        date = new.find("li", class_="cell012 tar fsn").text
        time = new.find("li", class_="cell024 tar fsn").text
        date_time = News.date_time_to_dateTime(date, time)
        # print("title :" + new_title)
        # print("url :" + new_url)
        # print("date :" + date)
        # print("time :" + time)
        # print("date_time :" + str(date_time))
        news_class = News( title= new_title, source=SOURCE, news_url=new_url,date_time=date_time) 
        all_news.append(news_class)

print("Total news count: ", len(all_news))
if(len(all_news) > 0):
    print("Example url: " + all_news[0].news_url)
        
#for news in newsTable:
```

```python
# for each news go 
for index,news in enumerate(all_news):
    url = news.news_url
    print("News_url: ", url)
    req = Request(url, headers={'User-Agent': 'Mozilla/5.0'})
    html = urlopen(req).read()
    soup = BeautifulSoup(html, 'html.parser')
    # try: 
    # news' header information
    header_element = soup.find(class_="news-content__inf").find("h2").text
    text = header_element + " "
    
    # paragraph_elements = soup.find_all("p")
    # for t in paragraph_elements:
    #     text += t.text + "\n"
        
    #check if it has news-content readingTime class if it has get the text
    content_element = soup.find(class_="news-content readingTime")
    if(content_element != None):
        text_elements = content_element.find_all("p")
        for t in text_elements:
            text += t.text + " "
    #check if it has gallerylist class if it has get the text
    content_element = soup.find(class_="gallery-list")
    if(content_element != None):
        text_elements = content_element.find_all("p")
        for t in text_elements:
            # sometimes the same text is duplicated because of the structure of the html, to prevent it check if it is the duplicated text
            if(t.find("p") != None): 
                continue
            print("text: ", t)
            text += t.text + " "
    
    
    print("Content: \n",text,sep="")
    print("\n")
    # except Exception as e:
    #     print("Error: ", e)
    #     print("Error in news: ", news.news_url)
    #     print("\n")
    all_news[index].content = text
    
    
```

```python
import csv
# save files in csv format
csv_filename = "bigpara_news.csv"
headers = ["title", "content", "date_time", "source", "news_url"]
# Open the CSV file in write mode
with open(csv_filename, mode="w", newline="", encoding="utf-8") as file:
    writer = csv.writer(file)
    
    # Write the headers
    writer.writerow(headers)
    
    # Write each news item as a row in the CSV
    for news in all_news[:1]:
        writer.writerow([
            news.title,
            news.content,
            news.date_time.strftime('%Y-%m-%d %H:%M') if news.date_time else "",  # Format datetime
            news.source,
            news.news_url
        ])
```

```python
import json

# todo: take this into class definition
def news_to_dict(news):
    return {
        "title": news.title,
        "content": news.content,
        "date_time": news.date_time.strftime('%Y-%m-%d %H:%M') if news.date_time else None,
        "source": news.source,
        "news_url": news.news_url
    }
    

json_filename = "bigpara_news.json"
# Convert each News object to a dictionary
news_dict_list = [news_to_dict(news) for news in all_news]

# Write to JSON file
with open(json_filename, "w", encoding="utf-8") as file:
    json.dump(news_dict_list, file, ensure_ascii=False, indent=4)
```
